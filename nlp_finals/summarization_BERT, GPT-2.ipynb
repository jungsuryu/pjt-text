```python
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
```

```python
stop_words = set(stopwords.words('english'))
```

```python
# 전처리 함수
from bs4 import BeautifulSoup
import urllib.request

def preprocess_sentence(sentence, remove_stopwords = True):
    # 불용어 제거 (Text)
    if remove_stopwords:
        tokens = ' '.join(word for word in sentence.split() if not word in stop_words if len(word) > 1)

    return tokens
```

```python
# Creation the list with new long block
max_length = 100  # minimum characters in each block
i = 0
bodies = []
while i < len(summary):
    body = ""
    body_empty = True
    while (len(body) < max_length) and (i < len(summary)):
        if body_empty:
            body = summary.iloc[i,0]
            body_empty = False
        else: body += " " + summary.iloc[i,0]
        i += 1
    bodies.append(body)
    print("Length of blocks =", len(body))
print(f"\nNumber of text blocks = {len(bodies)}\n")
print("Text blocks:\n", bodies)
```

# SBERT

```python
from sentence_transformers import SentenceTransformer, util
model = SentenceTransformer('all-MiniLM-L6-v2')
# model.to('cpu') 
```

```python
# BERT prediction into SBERT
pred = summary.loc[1,'bert_summary']
pred_vector = model.encode(pred)
pred_vector
```

```python
# calculate the similarity between question and answer choices
def cal_sim(i):
  pred = summary.loc[i,'bert_summary']
  pred_vector = model.encode(pred)

  ans = []
  for j in range(1,6):
    col = str(j)
    a = summary.loc[i,col]
    a_vector = model.encode(a)
    ans.append(a_vector)
  
  sim = []
  for an in ans:
    sim_score = pred_vector @ an
    sim.append(sim_score) 

  show = np.argmax(sim) + 1 
  return show
```

```python
answer_list = []

for i in range(len(summary)):
  answer = cal_sim(i)
  answer_list.append(answer)
```

```python
real_answer = summary.answer.values
```

```python
score = 0
length = len(summary)

for i in range(length):   
  if answer_list[i] == real_answer[i]:
    score += 1
```

```python
# BERTSum로 추출한 요약문 사용했을 때 정답률 계산
score/length
```
